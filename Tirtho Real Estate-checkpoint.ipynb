## scikit-learn-design

primarily three type of objects
1. Estimators - it estimates some parameter based on a dataset. Eg. Imputer`
   it has a fit and transform method. Fits the dataset and calculates internal parameters
   
3. Predictors - Linearregression is an example pf predictor.It has fit() and predict() method
   
5. Transformers - it takes input and returns output based on the learnings of the fit. it has a convinience function fit_transform()
   which fits and then transform. It also gives score which evaluates the predictions

## FEATURE SCALING

Primarily,two types of scailing methods
1 . MIn-max scailing (normalization)
    (value - min)/(max-min)
    SKlearn provides a class called Minmaxscaller for this
2 . Standarization
    (value - mean)/std
    Sklearn provides a class called standard Scaler for this

# PIPELINE

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy = 'median')),
    # ...... add as many as you want in your pipeline
    ('std_scaler', StandardScaler()),
])
```

```python
housing_num_tr = my_pipeline.fit_transform(housing)
```

```python
housing_num_tr
```

## EVALUATING THE MODEL

```python
from sklearn.metrics import mean_squared_error
housing_predictions = model.predict(housing_num_tr)
mse = mean_squared_error(housing_labels,housing_predictions)
rmse = np.sqrt(mse)
```

```python
mse
```

## USING BETTER EVALUATION TECHNIQUE - CROSS VALIDATION

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, housing_num_tr,housing_labels,scoring='neg_mean_squared_error',cv=10)
rmse_scores = np.sqrt(-scores)
```

```python
rmse_scores
```

```python
def print_scores(scores):
    print('Scores: ', scores)
    print("Mean: ",scores.mean())
    print("Standard Deviation : ", scores.std())
```

```python
print_scores(rmse_scores)
```

## TESTING THE MODEL ON TEST DATA

```python
x_test = strat_test_set.drop('MEDV',axis = 1)
y_test = strat_test_set["MEDV"].copy()
x_test_prepared = my_pipeline.transform(x_test)
final_predictions = model.predict(x_test_prepared)
final_mse = mean_squared_error(y_test,final_predictions)
final_rmse = np.sqrt(final_mse)
```
